# Classification model

In this final section of the notebook, we will evaluate three classifiers: `LogisticRegression`, `DecisionTree` and `RandomForest`. To select the best model, we will split the dataset into three partitions: training, validation and test.
- **Training set**: used to perform grid-search cross-validation and find the best model for each classifier.
- **Validation set**: used to evaluate the best model of each classifier and find the champion among the three of them.
- **Test set**: used to assess the generalization capabilities of the champion model.

## Logistic regression

For this model, we need to evaluate the main assumptions. 
- Outcome variable is categorical
- Observations are independent of each other
- No severe multicollinearity among X variables
- No extreme outliers
- Linear relationship between each X variable and the logit of the outcome variable
- Sufficiently large sample size

### Variable encoding
There are two categorical variables: `salary` and `department`. The variable `salary` is ordinal, therefore one-hot encoding is not suitable and we must apply a numerical encoding. Alternatively, we concluded from EDA that `accident` and `department`should be dropped for classification.

```python
# Copy data
df_encoded = data.copy()

# Encode salary
df_encoded['salary'].replace({'low': 0, 'medium': 1, 'high': 2}, inplace=True)

# Drop department and accident according to EDA conclusions
df_encoded.drop(labels=['department', 'accident'], axis=1, inplace=True)
```

### Outliers
Check for extreme outliers in the numerical variables and remove those observations in case they do not represent a big proportion of the dataset

```python
# Select variables to explore
vars_to_inspect = ['satisfaction_level', 'last_evaluation', 'average_monthly_hours', 'projects', 'tenure']

# Prepare figure
fig, axs = plt.subplots(len(vars_to_inspect), 1, figsize=(15, 10))

# Plot boxplots
for idx, var in enumerate(vars_to_inspect):
    sns.boxplot(data=df_encoded, x=var, ax=axs[idx])
    axs[idx].set_title(f'Boxplot of {var}')

# Layout
plt.tight_layout()
```

According to the boxplots above, only the variable `tenure` presents extreme outliers. In order to apply the `LogisticRegression` model, we are going to drop them in the next cell.

```python
# Calculate the quantiles
col = 'tenure'
p25 = df_encoded[col].quantile(q=0.25)
p75 = df_encoded[col].quantile(q=0.75)
IQR = p75 - p25

# Determine the number of outliers
num_outliers_upper = (df_encoded[col] > p75 + 1.5 * IQR).sum() 
num_outliers_lower = (df_encoded[col] < p25 - 1.5 * IQR).sum()
num_outliers       = num_outliers_lower + num_outliers_upper 

# Display
print(f'The number of outliers for the variable {col} is {num_outliers}')
print(f'This represents {round(num_outliers / len(df_encoded) * 100, 2)}% of the dataset')
```

```python
# Drop outliers
df_log = df_encoded[(df_encoded['tenure'] >= p25 - 1.5 * IQR) & ((df_encoded['tenure'] <= p75 + 1.5 * IQR))]
```

## Split into training, validation and test set
For the `LogisticRegression` model, we will use the data without outliers. Although for the `DecisionTree` and the `RandomForest` we could use the data with outliers, since these models are not affected by them, we will also use the data without outliers. We take this approach because the outliers only represented a ~6% of the dataset and this way the code results more neat. 

```python
# Separate features and target
X = df_log.drop(labels=['left'], axis=1)
y = df_log['left']

# Split into training and test sets
X_trval, X_test, y_trval, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_trval, y_trval, test_size=0.25, stratify=y_trval, random_state=42)

# Check dimensions
print(f'Shapes for training: X {X_train.shape} and y {y_train.shape}')
print(f'Shapes for validation: X {X_val.shape} and y {y_val.shape}')
print(f'Shapes for test: X {X_test.shape} and y {y_test.shape}')
```

## Grid-search cross-validation
Use `GridSearchCV` to find the best model for each of the three classifiers using the training set.

```python
%%time

# Instantiate the three classifiers
log    = LogisticRegression(max_iter=500, random_state=42)
tree   = DecisionTreeClassifier(random_state=42)
forest = RandomForestClassifier(random_state=42) 

# Hyperparameter grid for the three classifiers
params_log = {
    'penalty': ['l1','l2'],
    'C'      : [0.01, 0.1, 1, 10, 100],
    'solver' : ['saga']
}
params_tree = {
    'max_depth'        :[4, 6, 8, None], 
    'min_samples_leaf' : [2, 5, 1], 
    'min_samples_split': [2, 4, 6]
}
params_forest = {
    'max_depth'        : [3, 5], 
    'max_features'     : [1.0],
    'max_samples'      : [0.7, 1.0],
    'min_samples_leaf' : [1,2,3],
    'min_samples_split': [2,3,4],
    'n_estimators'     : [100, 500],
}

# Scoring metrics to calculate during cross-validation
scoring = {'accuracy', 'precision', 'recall', 'f1', 'roc_auc'}

# Fit a grid-search cross-validation for the three classifiers
gscv_log    = GridSearchCV(log, params_log, scoring=scoring, cv=4, refit='roc_auc').fit(X_train, y_train)
gscv_tree   = GridSearchCV(tree, params_tree, scoring=scoring, cv=4, refit='roc_auc').fit(X_train, y_train)
gscv_forest = GridSearchCV(forest, params_forest, scoring=scoring, cv=4, refit='roc_auc').fit(X_train, y_train)
```

Write a function to obtain the scores for the best model for each classifier to a `DataFrame`.

```python

def scores_to_df(gscv):
    '''
    Function to read the all the scores of the best model
    
    Inputs
    ------
    gscv  -- the fitted grid-search cross-validation object.
    
    Outputs
    -------
    df_scores -- a dataframe with the GSCV test scores for the best combination of hyperparameters.
    '''
    
    # Columns to read
    cols      = ['mean_test_' + metric for metric in scoring]
    df_gscv   = pd.DataFrame(gscv.cv_results_)
    
    # Get the scores from the best combination of hyperparameters
    df_scores = pd.DataFrame(df_gscv.loc[gscv.best_index_, cols]).transpose()
    
    # Rename for readability
    df_scores.rename(index={gscv.best_index_: str(gscv.estimator).split('(')[0]}, inplace=True)
    
    return df_scores
```

Write the scores for the three classifiers to a `DataFrame`.

```python
# Write scores to a dataframe
training_scores = pd.concat([scores_to_df(gscv_log), scores_to_df(gscv_tree), scores_to_df(gscv_forest)], axis=0)
training_scores
```

## Validation
Use the validation set to evaluate the three models resulting from `GridSearchCV` and select a champion.

```python
# Preallocate list to save results from the three models
validation_list = list()

# Loop models
for model in [gscv_log, gscv_tree, gscv_forest]:

    # Estimate predictions on the validation set
    y_pred = model.predict(X_val)
    
    # Prepare output
    validation_score = dict()
    
    # Loop metrics
    for score_name, score_method in zip(scoring, [accuracy_score, precision_score, recall_score, f1_score, roc_auc_score]):
        
        # Add scores to a dictionary
        validation_score[score_name] = score_method(y_val, y_pred)
        
    # Write to df
    validation_score_df = pd.DataFrame(validation_score, index=[str(model.estimator).split('(')[0]])
    
    # Add to list
    validation_list.append(validation_score_df)
        
# Concatenate into a single dataframe
validation_scores = pd.concat(validation_list, axis=0)
validation_scores
```

According to these validation results: 

The `LogisticRegression` classifier does not meet the minimum expectations, since all the scores aside from accuracy are very low. These poor results suggest the classifier may have not converged with the hyperparameters explored. 

The `DecisionTree` classifier yielded excellent performance (very close to the performance during the grid-search cross-validation). All the performance metrics are above 0.93.

The `RandomForest` classifier also obtained excellent performance. In this case, the performance is slightly superior than the one from the tree, what makes this model, the champion.

```python
# Concatenate training + validation sets
X_final = pd.concat([X_train, X_val], axis=0)
y_final = pd.concat([y_train, y_val], axis=0)

# Select champion classifier
champion = gscv_forest.best_estimator_

# Fit the champion model to the concatenated set
champion.fit(X_final, y_final)
```

## Test
Finally, evaluate the re trained champion classifier on the test set, which was heldt out during all the previous stages.

```python
# Estimate predictions on the validation set
y_pred = champion.predict(X_test)

# Prepare output
test_score = dict()

# Loop metrics
for score_name, score_method in zip(scoring, [accuracy_score, precision_score, recall_score, f1_score, roc_auc_score]):

    # Add scores to a dictionary
    test_score[score_name] = score_method(y_test, y_pred)

# Write to df
test_score = pd.DataFrame(test_score, index=[str(champion).split('(')[0]])
test_score
```

The results of **the champion classifier on the test set are also excellent, as in the training and validation** stages. This indicates the model implemented is robust. The `RandomForest` classifier was able to very accurately classify which employees left and stayed in the company. 

For the final step in this notebook, let us inspect the feature importances assigned by the champion classifier. Recall feature importance is a measure of the decrease in `Gini Entropy` and therefore, the discriminability of the features utilized by the `RandomForest` classifier. 

```python
# Get feature importances
importances = champion.feature_importances_

# Get indices of top features
top = np.argpartition(champion.feature_importances_, -5)[-5:]

# Get column labels of top 10 features 
features = X.columns[top]

# Filter feature importances to consist of top 10 feature importances
importances = importances[top]

# Write importances to dataframe
importances_df = pd.DataFrame({'Feature': features, 'Importance': importances}).sort_values('Importance')
fig, ax        = plt.subplots(1, 1, figsize=(10, 5))

# Plot
importances_df.plot(kind='barh', x='Feature', y='Importance', ax=ax)

# Layout
ax.set_title('Random Forest: Feature Importances for Employee Leaving', fontsize=12)
ax.set_ylabel('Feature')
ax.set_xlabel('Importance')

plt.show()
```

Not surprisingly, the `RandomForest` selected the `satisfaction_level` as the most discriminative feature. Other features which were related to employees leaving during EDA, like `tenure`, the number of `projects`, and the `last_evaluation` were also assigned high importance.
